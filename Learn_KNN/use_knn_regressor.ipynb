<h2>Import Tools</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from model import (
    KN_Classifier,
    KN_Regressor,
    z_score_normalize,
    train_test_split_manual,
    train_test_split_manual_p
)
```

<h2> Un dataset atypique : A l'aide de np.random + Visualisations</h2>

```python
np.random.seed(42)
m = 100 #Creationg 100 samples
X = np.linspace(0,10,m).reshape(m,1) #Generation des donnes d'exemples
def f(X)  :
    return 2 * X + 1 + np.random.randn(m,1) # ax+b plus le bruit

y = f(X)
plt.scatter(X,y,c='g')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split_manual(X, y, test_size=0.2)

model = KN_Regressor(k=3)
model.fit(X_train,y_train)

print(model.score(X_test,y_test))
y_pred = model.predict(X)

plt.scatter(X,y,c='black')
plt.plot(X,y_pred,lw='3',c='r')
```
