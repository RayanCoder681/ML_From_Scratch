# Utilisation de l'algorithme de K-Nearest Neighbords Classifier

<h2>Import Tools</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from model import (
    KN_Classifier,
    accuracy_score,
    train_test_split_manual_p,
)
```

<h2>Chargement des donnees</h2>

```python
iris = pd.read_csv('./datasets/iris.csv')
print(iris.shape)
iris = iris.drop('Id',axis=1)
iris.head()
```

<h2>Pre-Processing</h2>

```python

iris = iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']]
iris.dropna(axis=0,inplace=True)
iris['Species']=iris['Species'].map({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
print(iris.shape)
print(iris.info())
iris.head()
```

```python
y = iris['Species']
X = iris.drop('Species',axis=1)

plt.scatter(X['SepalLengthCm'],X['SepalWidthCm'],c=y)
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
```

<h2>Define and Train the Model</h2>

```python
randomState = 42
X_train,X_test,y_train,y_test = train_test_split_manual_p(X,y,0.2,randomState)

model = KN_Classifier(k=3)

model.fit(X_train,y_train)
 
print(f"Le model obtient un score de : {100*model.score(X_test,y_test)}% ")

y_pred = model.predict(X_test)

print(f"La precision globale du model est de : {100*accuracy_score(y_test,y_pred)}%")
```
